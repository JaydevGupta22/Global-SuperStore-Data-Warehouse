```python
df.columns = [c.strip().lower().replace(' ','_') for c in df.columns]
```

```python
missing_postal = df['postal_code'].isna().sum()
print(f"Missing postal codes: {missing_postal}")
```

```python
df[df['postal_code'].isna()][['country','state','city','region','postal_code']].head(10)
```

```python
dim_customer = df[['customer_id','customer_name','segment']].drop_duplicates().reset_index(drop=True)

# create surrogate key if you want (recommended for warehouse)
dim_customer.insert(0, 'customer_sk', range(1, len(dim_customer)+1))
```

```python
# Drop rows with missing postal codes
df = df.dropna(subset=['postal_code'])

# Confirm removal
print("Missing postal codes after cleaning:", df['postal_code'].isna().sum())
```

```python
# Remove duplicates
df = df.drop_duplicates()

print("Remaining duplicates:", df.duplicated().sum())
```

```python
dim_customer = df[['customer_id', 'customer_name', 'segment']].drop_duplicates().reset_index(drop=True)
dim_customer['customer_sk'] = dim_customer.index + 1
```

```python
dim_customer
```

```python
dim_product = df[['product_id', 'category', 'sub-category', 'product_name']].drop_duplicates().reset_index(drop=True)
dim_product['product_sk'] = dim_product.index + 1
```

```python
dim_product
```

```python
dim_region = df[['country', 'state', 'city', 'postal_code', 'region']].drop_duplicates().reset_index(drop=True)
dim_region['region_sk'] = dim_region.index + 1
```

```python
dim_region
```

```python
dim_ship_mode = df[['ship_mode']].drop_duplicates().reset_index(drop=True)
dim_ship_mode['ship_mode_sk'] = dim_ship_mode.index + 1
```

```python
dim_ship_mode
```

```python
order_dates = df[['order_date']].rename(columns={'order_date':'full_date'})
ship_dates = df[['ship_date']].rename(columns={'ship_date':'full_date'})

dim_date = pd.concat([order_dates, ship_dates]).drop_duplicates().reset_index(drop=True)
dim_date['date_sk'] = dim_date.index + 1
```

```python
dim_date
```

```python
# Merge to get SKs
fact_sales = df.merge(dim_customer, on='customer_id') \
               .merge(dim_product, on='product_id') \
               .merge(dim_region, on=['country','state','city','postal_code','region']) \
               .merge(dim_ship_mode, on='ship_mode') \
               .merge(dim_date.rename(columns={'full_date':'order_date', 'date_sk':'order_date_sk'}), on='order_date', how='left') \
               .merge(dim_date.rename(columns={'full_date':'ship_date', 'date_sk':'ship_date_sk'}), on='ship_date', how='left')

# Select only the relevant fact columns
fact_sales = fact_sales[['customer_sk','product_sk','region_sk','ship_mode_sk','order_date_sk','ship_date_sk','sales']]
```
